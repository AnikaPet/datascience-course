```python
import numpy as np
import pandas as pd 
```

```python
# Requirements for dataset: datetime column, at least 3 string columns and at least 3 numeric columns.
# Dataset(source: Kaggle) from 2020. about salaries in IT industry in Europe
df = pd.read_csv("../IT Salary Survey EU  2020.csv")
```

```python
# Getting to know dataset and available data types and formats.

df.info()
```

```python
# Manipulate datetime data, extract year, month and day as separete columns.

#Timestamp is datatype object, add Timestamp1 column as datetime
df['Timestamp1']=pd.to_datetime(df['Timestamp'], format = '%d/%m/%Y %H:%M:%S')
```

```python
# Extract year, month and day as separate columns.

#df['Year']=df['Timestamp1'].dt.year
#df['Month']=df['Timestamp1'].dt.month
df['Day']=df['Timestamp1'].dt.day
```

```python
# Check if new columns are there.

df
```

```python
# We like %Y-%m-%d format better, so let's overwrite Timestamp1 column and remove time.

df['Timestamp']=df['Timestamp1'].dt.strftime('%Y-%m-%d')
```

```python
# Convert to datetime type.
df['Timestamp']=pd.to_datetime(df['Timestamp'], format = '%Y-%m-%d')
```

```python
df['Timestamp']
```

```python
# Let's apply some string functions to clean data. 

df['Other technologies/programming languages you use often'].unique()

# We want to do some string formatting for this column. Let's replace / with , and remove unnecessary blank spaces.
```

```python
df['Other technologies/programming languages you use often']=df['Other technologies/programming languages you use often'].str.replace('/',',').replace(' ,',',')
```

```python
# We can also remove ;
df['Other technologies/programming languages you use often']
```

```python
df['Other technologies/programming languages you use often']=df['Other technologies/programming languages you use often'].str.replace(';',',')
```

```python
# Looks better!
df['Other technologies/programming languages you use often']
```

```python
# We can also standardize Company size column. 

df['Company size'].unique()
```

```python
# Two different ways.
#df['Company size']=df['Company size'].str.replace('up to 10','0-10').replace('1000+','1000-9999')

df.replace({
    'Company size': {
        'up to 10': '0-10',
        '1000+': '1000-9999',
    }
},inplace=True)
```

```python
df['Company size'].unique()
```

```python
# Let's add new column as math interval so the representaion is better.
df['CompanySizeMin','CompanySizeMax']=df['Company size'].str.split('-')
```

```python
# Explore more string columns. We can see value 'Divers' as interesting point in column Gender.
df['Gender'].unique()
```

```python
# Let's clean up column Seniority level.
df['Seniority level'] = df['Seniority level'].str.upper()
#df['Company type'] = df['Company type'].str.lower()
```

```python
df['Seniority level'].unique()
```

```python
# For better filtering, let's remove NaNs and non-numeric values.
#df['Total years of experience'].fillna('NO NOTE', inplace = True)

df.replace({
    'Total years of experience': {
        '1 (as QA Engineer) / 11 in total': '11',
        '6 (not as a data scientist, but as a lab scientist)': '6',
        '15, thereof 8 as CTO':'15',
        'less than year': '1'
    }
},inplace=True)


#df['Total years of experience']=df['Total years of experience'].replace('15, thereof 8 as CTO','15').replace('less than year','1')
```
